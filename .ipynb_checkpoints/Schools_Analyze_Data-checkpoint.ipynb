```python
def plot_confusion_matrix(cm, classes=[0, 1], title='confusion matrix',
                          normalize=False,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                horizontalalignment="center", size=20,
                color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
def my_classifier(X,y,estimator, best_param, fig, ax):
    
    pipeline = [('scaler', StandardScaler())]
    
    if estimator == 'logistic':
        pipeline.append( ('estimator',  LogisticRegressionCV(Cs=10, cv=5, random_state=1)))
    
    elif estimator == 'naive_bayes':
        pipeline.append(('estimator',  naive_bayes.GaussianNB()))
    
    elif estimator == 'k_nearest_neighbors':
        pipeline.append(('estimator',  KNeighborsClassifier(n_neighbors = 9)))
    elif estimator == 'decision_tree':
        pipeline.append(('estimator', DecisionTreeClassifier()))            
    elif estimator == 'SVM':
        pipeline.append(('estimator', svm.SVC(kernel='rbf', probability = True))) #other kernel options 'rbf', 'poly', 'sigmoid'               
    elif estimator == 'random_forest':        
        pipeline.append(('estimator', RandomForestClassifier(n_estimators = 300)))
    elif estimator == 'gradient_boosting':
        pipeline.append(('estimator', GradientBoostingClassifier(n_estimators = 300)))
        
    elif estimator == 'extreme_gb':
        pipeline.append(('estimator', XGBClassifier(n_estimators = 300, learning_rate = 0.1)  ))
    
    elif estimator == 'neural_MLP':
        pipeline.append(('estimator', MLPClassifier(
        hidden_layer_sizes=(100,),
        solver='adam',
        verbose=False,
        activation='tanh',
        learning_rate='adaptive',
        early_stopping=False,
        max_iter=200,
        batch_size=32)))
    
    else:
        print('dont know that estimator')
        return None    
                        
    pipeline = Pipeline(pipeline)
                        
    #stuff to save                
    acc_test=[]
    roc_auc= []
    precision = []
    recall = []
    neighbors=[]
     
        
    y_test_score = 0
    max_score = 0
    
    saved_X_test = []
    saved_X_train = []
    saved_y_test = []
    saved_y_train = []
    saved_y_proba = []

    kf = KFold(n_splits=5, shuffle = True)
    kf.get_n_splits(X)
    
    for train_index, test_index in kf.split(X):
        
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        pipeline.fit(X_train, y_train)
                            
        y_proba = pipeline.predict_proba(X_test)[:, 1]
        y_pred = pipeline.predict(X_test)            
        
        y_test_score = accuracy_score(y_test, y_pred)        
        acc_test.append(y_test_score)
        precision.append(precision_score(y_test, y_pred))
        recall.append(recall_score(y_test, y_pred))
        roc_auc.append(roc_auc_score(y_test, y_proba))

        
        if y_test_score > max_score:
            max_score = y_test_score
            saved_X_test = X_test
            saved_X_train = X_train
            saved_y_test = y_test
            saved_y_train = y_train
            saved_y_proba = y_proba
            saved_pipeline = pipeline.fit(X_train, y_train)
            saved_roc_score = roc_auc_score(y_test, y_proba)
            
            if estimator == 'logistic':
                saved_coef = pipeline.steps[-1][1].coef_             
    
    x,y,z = roc_curve(y_test, y_proba)    
    label = '%s, %.1f%%' % (estimator, 100*roc_auc_score(y_test, y_proba) )
    fig.suptitle('ROC CURVES', fontweight='bold');
    ax.plot([0, 1], [0, 1], color='k', linestyle='--', linewidth=.4);
    ax.plot(x,y, label=label);
    ax.axis('equal');
    ax.set(xlim=[-.02, 1.02], ylim=[-.02, 1.02]);
    ax.legend(loc='best');
    ax.grid(True);
    
    
    print('best classification report')
    print(classification_report(saved_y_test, saved_pipeline.predict(saved_X_test)))
#    plot_confusion_matrix(confusion_matrix(saved_y_test, saved_pipeline.predict(saved_X_test)))
    

    print('av_roc_auc', np.mean(roc_auc))  
    print('av_acc_test', np.mean(acc_test))
    print('av_precision', np.mean(precision))
    print('av_recall', np.mean(recall))

        
    if estimator == 'logistic':
        return np.mean(acc_test), np.mean(roc_auc), np.mean(precision), np.mean(recall), saved_coef, fig, ax                
    else:
        return np.mean(acc_test), np.mean(roc_auc), np.mean(precision), np.mean(recall), None, fig, ax
```

```python
#first find best neighbor before using pipeline
def best_neighbor(X,y):
    kf = KFold(n_splits=5, shuffle = True)
    kf.get_n_splits(X)

    n_neighbors = []
    for train_index, test_index in kf.split(X):
    
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        param_grid = {'n_neighbors': range(1, 10)}

        grid = GridSearchCV(KNeighborsClassifier(), 
                    param_grid, 
                    cv=3, 
                    scoring='accuracy')

        grid = grid.fit(X_train, y_train)
        n_neighbors.append(grid.best_params_['n_neighbors'])
    return int(np.mean(n_neighbors))

best_neighbor = best_neighbor(X,y)
print(best_neighbor)

#now use pipeline
acc_test, roc_score, precision, recall, neighbors =  my_classifier(X,y, 'k_nearest_neighbors',best_neighbor)
```

```python
#first find best parameters for RandomForest before using pipeline
def best_params(X,y):
    kf = KFold(n_splits=5, shuffle = True)
    kf.get_n_splits(X)

    n_estimators = []
    for train_index, test_index in kf.split(X):
    
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        param_grid = {'n_estimators': [100, 200, 300, 400], 'n_jobs': [-1]}

        grid = GridSearchCV(RandomForestClassifier(), 
                    param_grid, 
                    cv=3, 
                    scoring='accuracy')

        grid = grid.fit(X_train, y_train)

        print(grid.best_params_, grid.best_estimator_)
        n_estimators.append(grid.best_params_['n_estimators'])
        
    return mode(n_estimators)[0][0]

best_parameter = best_params(X,y)
print(best_parameter)

#now use pipeline
acc_test, roc_score, precision, recall, _ =  my_classifier(X,y, 'randomforest',best_parameter)
```

```python
#this function tries to predict schools that are in a given percentile of a given variable

def find_schools(df, feature, percentile, best_or_worst):
#make a flag for being in a certain percentile of a given variable
    X = df[['size','retention','branches', 'online_only', 'under_investigation', 'most_common_degree', 'highest_degree',
       'faculty_salary', 'instructional_expenditure_per_fte',
       'tuition_revenue_per_fte', 'part_time_share',
       'age_entry', 'percent_dependent', 'first_generation', 'percent_black',
        'avg_family_income','ind_low_income', 'dep_low_income', 'loan_principal',
       'federal_loan_rate', 'students_with_any_loans',
       'pell_grant_debt', 'percent_pell_grant',
       'fafsa_sent', '7_yr_repayment_completion', '5_year_declining_balance',
       'relig_y_n', 'accred_y_n', 'retention_listed_y_n',
       'fac_salary_listed_y_n', '7_yr_repayment_completion_y_n',
       '5_year_declining_balance_y_n', 'for_profit']]

    X['flag'] = 0
    for school_num in X.index:
        if best_or_worst == 'best':
            if X[feature][school_num] > np.percentile(df[feature], percentile ):
                X = X.set_value(school_num, 'flag', 1)
        elif best_or_worst == 'worst':
            if X[feature][school_num] < np.percentile(df[feature], percentile ):
                X = X.set_value(school_num, 'flag', 1)
        else:
            print('you must choose best or worst')
            
    y = X.pop('flag')

    X_new = SelectKBest(chi2, k=2).fit(X, y)

    important_features = []
    for i, column in enumerate(X.columns):
        if X_new.pvalues_[i] < 0.05:
            important_features.append(column)
    
    X = df[important_features]
    del X[feature]

    my_classifier(X,y, 'extreme_gb', 300, _, _)

    schools = []
    for school_num in df.index:
        if best_or_worst == 'best':
            if df[feature][school_num] > np.percentile(df[feature], percentile):
                schools.append((df[feature][school_num],df['name'][school_num]))
                schools.sort(reverse = True)
        else:
            if df[feature][school_num] < np.percentile(df[feature], percentile):
                schools.append((df[feature][school_num],df['name'][school_num]))
                schools.sort()
    
    for school in schools:
        print('%.3f, %s' % (school[0], school[1]))
```

```python
#condense multibranch schools to one listing for cleaner D3 visualization
condensed_df = df.groupby(['ope6_id']).agg({"name": "first", 
                             "for_profit": "max",
                             'tuition_revenue_per_fte': "mean",
                             "branches": "max",
                             'instructional_expenditure_per_fte': "mean",
                             "faculty_salary": "mean",
                             "under_investigation": "max",
                              "5_year_declining_balance": "mean",
                              "7_yr_repayment_completion": "mean",              
                               "fafsa_sent": "mean",
                               "percent_pell_grant": "mean"})

#get rid of hyphens in name for D3
def drop_hyphens(name):
    return name.split('-')[0]


condensed_df['name'] = condensed_df['name'].apply(drop_hyphens)
condensed_df['for_profit'].replace(to_replace = [0,1], value = ['non profit', 'for profit'], inplace = True)
condensed_df['under_investigation'].replace(to_replace = [0,1], value = ['no', 'yes'], inplace = True)
```

```python
#combine all Everest College entries into one
everest = pd.DataFrame({'name': ['Everest College'],
                         "for_profit": ["for profit"],
                             'tuition_revenue_per_fte':  [condensed_df[condensed_df['name'].str.contains('Everest')][ 'tuition_revenue_per_fte'].mean()],
                             "branches": [condensed_df[condensed_df['name'].str.contains('Everest')][ 'branches'].sum()],
                             'instructional_expenditure_per_fte': [condensed_df[condensed_df['name'].str.contains('Everest')][ 'instructional_expenditure_per_fte'].mean()],
                             "faculty_salary": [condensed_df[condensed_df['name'].str.contains('Everest')][ 'faculty_salary'].mean()],
                             "under_investigation": ["no"],
                              "5_year_declining_balance": [condensed_df[condensed_df['name'].str.contains('Everest')][ '5_year_declining_balance'].mean()],
                         "7_yr_repayment_completion": [condensed_df[condensed_df['name'].str.contains('Everest')][ '7_yr_repayment_completion'].mean()],
                               "fafsa_sent": [condensed_df[condensed_df['name'].str.contains('Everest')][ 'fafsa_sent'].mean()],
                               "percent_pell_grant": [condensed_df[condensed_df['name'].str.contains('Everest')][ 'percent_pell_grant'].mean()],
                                "type": ['for profit']})
cols = ['name', 'for_profit', 'tuition_revenue_per_fte', 'branches', 'instructional_expenditure_per_fte', 'faculty_salary', 'under_investigation', '5_year_declining_balance', '7_yr_repayment_completion','fafsa_sent', 'percent_pell_grant']
everest = everest[cols]
condensed_df = condensed_df[~condensed_df['name'].str.contains('Everest')]
condensed_df = pd.concat([condensed_df,everest])

```

```python
#make an extra column for four types for D3 plotting
condensed_df['type'] = ''
for school_num in condensed_df.index:
    if condensed_df['for_profit'][school_num] == 'non profit' and condensed_df['under_investigation'][school_num] == 'yes':
         condensed_df = condensed_df.set_value(school_num, 'type', 'non profit under investigation')
    elif condensed_df['for_profit'][school_num] == 'non profit' and condensed_df['under_investigation'][school_num] == 'no':
         condensed_df = condensed_df.set_value(school_num, 'type', 'non profit')
    elif condensed_df['for_profit'][school_num] == 'for profit' and condensed_df['under_investigation'][school_num] == 'yes':
         condensed_df = condensed_df.set_value(school_num, 'type', 'for profit under investigation')
    else:
         condensed_df = condensed_df.set_value(school_num, 'type', 'for profit')
condensed_df.to_csv(path_or_buf = 'myschools.csv', index = False, header = ['schoolname', 'for profit', 'tuition revenue per fte', 'branches', 'instructional expenditure per fte', 'faculty salary', 'under investigation','5 year declining balance', '7_yr_repayment_completion','percent fafsa', 'percent pell grant','type' ])
```
